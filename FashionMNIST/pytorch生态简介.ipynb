# pytorch生态简介

## torchvision

**包含了计算机视觉中常用的数据集、模型和图像处理的方式**<br>
datasets:包含了计算机视觉中常见的数据集<br>
transforms:包含了常见的数据增强方法，缩小、放大、水平或垂直翻转等<br>
models:提供了一些预训练好的模型供我们使用<br>
io:提供了视频、图片和文件的IO操作的功能，包含读取、写入、编码和解码处理操作<br>
ops:提供了计算机视觉的特定操作，如NMS\ROIAlign\RoIPool等方法<br>
utils:提供了一些可视化的方法

## PyTorchVideo

**提供了加速视频理解研究所需的模块化和高效的API。支持不同的深度学习视频组件，如视频模型、视频数据集和视频特定转换。也提供了modelzoo,可使用先进的预训练模型及其评判标准**

**主要亮点**
基于pytorch:使所有pytorch生态系统组件的使用变得容易<br>
Model Zoo:提供了较多SOTA模型，简化了模型调用<br>
数据预处理和常见数据：支持地应数据集的数据预处理，还支持randaug,augmix等数据增强trick<br>
模块化设计:类似于torchvision,提供许多模块方便用户调用修改，在torchvideo中具体包括data, transforms, layer,model,accelerator等模块，方便用户调用和读取<br>
支持多模态:对多模态的支持包括了visual和audio,未来会支持更多模态<br>
移动端部署优化：支持对移动端的部署优化
